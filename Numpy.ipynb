```python
#reshape

a=np.arange(48)
print(a)
print(a.ndim)
b=a.reshape(8,6)
print(b)
```

```python
a=np.arange(48)
print(a)
print(a.ndim)
b=a.reshape(12,4)
print(b)
```

```python
a=np.arange(48)
print(a)
print(a.ndim)
b=a.reshape(6,-1)
print(b)
```

```python
#transpose
a=np.arange(48)
b=a.reshape(6,-1)
c=b.T
print(b)
print(c)
```

```python
#flatten
a=np.arange(48)
b=a.reshape(6,-1)
c=b.T
print(c)
print(c.flatten()) #used in deep learning
```

```python
#arithmetic operator
a=np.array([3,1,2,6])
b=np.array([0,1,3,4])
print(a+b)
print(a-b)
print(a*b)
print(a>2)
```

```python
a=np.array([[3,6,5],[8,3,6]])
b=np.array([[7,4,9],[9,3,8]])
print('a is',a)
print('b is',b)
print(a+b)
print(a-b)
print(a*b)
print(a>2)
```

```python
#dot product
x=np.linspace(1,10,8).reshape(4,2)
print(x)
y=np.linspace(1,10,10).reshape(2,5)
print(y)
z=x.dot(y)
print(z)
print(z.shape)
```

```python
x=np.linspace(1,10,8).reshape(4,2,)
y=np.linspace(1,10,10).reshape(2,5)
z=x.dot(y)
print(z)
print(z.shape)
```

```python
x=np.linspace(1,10,8).reshape(4,2,)
y=np.linspace(1,10,10).reshape(2,5)
z=x.dot(y)
print(z)
print(z.min())
print(z.max())
print(z.sum())
print(z.sum(axis=1)) #row wise sum
print(z.sum(axis=0)) #column wise sum
```

```python
#statistical function
x=np.linspace(1,10,8).reshape(4,2)
y=np.linspace(1,10,10).reshape(2,5)
z=x.dot(y)
print(z)
print(np.mean(z))
print(np.var(z))
print(np.median(z))
```

```python
#Indexing Slicing
x=np.linspace(1,10,8)
print(x)
print(x[5])
print(x[2::2])
```
