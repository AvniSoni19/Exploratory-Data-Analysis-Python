```python
week_day=[1,2,3,4,5,6,7]
temp=[50,51,52,58,47,47,46]

plt.plot(week_day, temp) #plot(x,y)

plt.xlabel('Weeks') #label_x
plt.ylabel('Temperature') #Label_y

plt.title('Weather Graph') #title
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,53,54,55,58]
min_t=[39,22,21,26,32,35,18]
avg_t=[50,50,50,50,50,50,50]
```

```python
plt.plot(days,max_t,'o--g') #Format string fmt_str=[marker'o'][linestyle'--'][color'g'] in any order
plt.plot(days,min_t,'o--b')
plt.plot(days,avg_t,'o--y')
plt.xlabel('Week')
plt.ylabel('Temperature')
plt.title("Weather Graph")
```

```python
#marker size
plt.plot(days,max_t,'o--g',markersize=10)
```

```python
#legend function
plt.plot(days,max_t,label='max') 
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.xlabel('Week')
plt.ylabel('Temperature')
plt.title("Weather Graph")
plt.legend(loc='best',fontsize='small')
plt.grid(which='both')
```

Subplot

```python
year=[2014,2015,2016,2017]
income=[28000,45000,50000,46000]
expense=[2800,3000,2800,3400]

plt.subplots_adjust(hspace=0.4,wspace=0.7)

plt.subplot(1,4,1) # (total row, total column, subplot serial number)
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(1,4,2)
plt.bar(year,expense,color='c')
plt.xlabel('abc')

plt.subplot(1,4,3)
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(1,4,4)
plt.bar(year,expense,color='c')
plt.xlabel('abc')
```

```python
year=[2014,2015,2016,2017]
income=[28000,45000,50000,46000]
expense=[2800,3000,2800,3400]

plt.subplots_adjust(hspace=0.010,wspace=0.5)

plt.subplot(2,2,1) 
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(2,2,2)
plt.bar(year,expense,color='c')
plt.xlabel('abc')

plt.subplot(2,2,3)
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(2,2,4)
plt.bar(year,expense,color='c')
plt.xlabel('abc')
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
embarked_group=df.groupby('Embarked').agg(Total_Records=('Embarked','size'))
print(embarked_group.index.values)
print(embarked_group['Total_Records'].values)

plt.bar(x=embarked_group.index.values,height=embarked_group['Total_Records'].values,color='pink')
plt.xlabel("Embarked")
plt.ylabel('Count')
plt.title("_________Embarked Count________")
plt.show()
```

```python
embarked_group=df.groupby('Embarked').agg(Total_Records=('Embarked','size'))
embarked_group
```

```python
embarked_group=df.groupby('Embarked').agg(Fare_collected=('Fare','sum'))
print(embarked_group.index.values)
print(embarked_group['Fare_collected'].values)

plt.pie(labels=embarked_group.index.values,x=embarked_group['Fare_collected'].values,
        shadow=True,
        explode=[0.2,0,0,],
        counterclock=False,
        colors=['b','y','g'],
        autopct='%1.2f%%')

plt.title("_________Pie Chart________")
plt.show()
```

```python
#countplot
sns.countplot(df['Survived'])
plt.xlabel("Column Survived")
plt.ylabel("Count -->")
plt.title("SNS Countplot")
plt.show()
```

```python
sns.swarmplot(df['Embarked'],df['Fare'],palette=['orange','yellow','cyan'])
```
