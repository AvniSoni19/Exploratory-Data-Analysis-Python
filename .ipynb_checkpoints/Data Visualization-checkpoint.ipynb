```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
week_day=[1,2,3,4,5,6,7]
temp=[50,51,52,58,47,47,46]

plt.plot(week_day, temp) #plot(x,y)

plt.xlabel('Weeks') #label_x
plt.ylabel('Temperature') #Label_y

plt.title('Weather Graph') #title
```

```python
plt.plot(week_day, temp, color='red')
```

```python
"""
color
'b' -> blue
'g' -> green
'r' -> red
'c' -> cyan
'm' -> magenta
'y' -> yellow
'k' -> black
'w' -> white
"""
```

```python
plt.plot(week_day, temp, color='g', linestyle=':')
```

```python
#marker
plt.plot(week_day, temp, color='g', linestyle=':', marker='|')
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,53,54,55,58]
min_t=[39,22,21,26,32,35,18]
avg_t=[50,50,50,50,50,50,50]
```

```python
plt.plot(days,max_t,'o--g') #Format string fmt_str=[marker'o'][linestyle'--'][color'g'] in any order
plt.plot(days,min_t,'o--b')
plt.plot(days,avg_t,'o--y')
plt.xlabel('Week')
plt.ylabel('Temperature')
plt.title("Weather Graph")
```

```python
#marker size
plt.plot(days,max_t,'o--g',markersize=10)
```

```python
#legend function
plt.plot(days,max_t,label='max') 
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.xlabel('Week')
plt.ylabel('Temperature')
plt.title("Weather Graph")
plt.legend(loc='best',fontsize='small')
plt.grid(which='both')
```

Barplot

```python
company=['Google','Amazon','MSFT','Facebook']
revenue=[90,136,89,27]

plt.bar(company,revenue)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
```

```python
#Horizontal Bar
company=['Google','Amazon','MSFT','Facebook']
revenue=[90,136,89,27]

plt.barh(company,revenue)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
```

```python
company=['Google','Amazon','MSFT','Facebook']
revenue=[90,136,89,67]
profit=[56,93,87,54]

plt.bar(company,revenue,label='revenue')
plt.bar(company,profit,label='profit')
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.legend(loc='best',fontsize='small')
```

Subplot

```python
year=[2014,2015,2016,2017]
income=[28000,45000,50000,46000]
expense=[2800,3000,2800,3400]

plt.subplots_adjust(hspace=0.4,wspace=0.7)

plt.subplot(1,4,1) # (total row, total column, subplot serial number)
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(1,4,2)
plt.bar(year,expense,color='c')
plt.xlabel('abc')

plt.subplot(1,4,3)
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(1,4,4)
plt.bar(year,expense,color='c')
plt.xlabel('abc')
```

```python
year=[2014,2015,2016,2017]
income=[28000,45000,50000,46000]
expense=[2800,3000,2800,3400]

plt.subplots_adjust(hspace=0.010,wspace=0.5)

plt.subplot(2,2,1) 
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(2,2,2)
plt.bar(year,expense,color='c')
plt.xlabel('abc')

plt.subplot(2,2,3)
plt.bar(year,income,color='y')
plt.xlabel('xyz')

plt.subplot(2,2,4)
plt.bar(year,expense,color='c')
plt.xlabel('abc')
```
