```python
attrition=[1233,237]
attrition_labels=['Yes','No']
plt.pie(x=attrition, labels=attrition_labels, radius=1, autopct='%0.1f%%',explode=[0,0.1])
plt.title('Attrition')
```

```python
x=df['Attrition']
y=df['Age']
plt.scatter(x,y,label='Attrition on ages', color='g',marker='o')
plt.xlabel('Attrition')
plt.ylabel('Ages')
plt.title('Title')
plt.legend()
```

```python
sns.catplot(x='Gender',y='Age',data=df,kind='swarm',color='g')
```

```python
sns.swarmplot(x=df['Attrition'],y=df['DailyRate'],palette=['yellow','black'])
```
