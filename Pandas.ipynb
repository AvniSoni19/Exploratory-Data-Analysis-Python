```python
us_weather=pd.DataFrame({'City':['New York','Chicago','Orlando'],'Temperature':[21,14,35],'Humidity':[68,65,75]})
us_weather
```

```python
india_weather=pd.DataFrame({'City':['Mumbai','Delhi','Bangalore'],'Temperature':[32,45,30],'abc':[1,4,3],'Humidity':[80,60,78]})
india_weather
```

```python
df_new=pd.concat([india_weather,us_weather]).reset_index(drop=True)
df_new
```

```python
df_new=pd.concat([india_weather,us_weather],ignore_index=True) #Column wise concatenation
df_new
```

```python
df_new=pd.concat([india_weather,us_weather],axis=1) #Row wise concatenation
df_new1=pd.DataFrame(df_new,index=[1,5,7])
df_new1
```

Changing Index

```python
df_new=pd.concat([india_weather,us_weather])
df_new.index=[1,2,3,4,5,6]
df_new
```

Merge

```python
df_1=pd.DataFrame({'City':['New York','Bangalore','Chicago','Orlando'],'Humidity':[68,58,65,75]})
df_1
```

```python
df_2=pd.DataFrame({'City':['Mumbai','Delhi','Chicago','Bangalore'],'Humidity':[80,60,50,78]})
df_2
```

```python
df_new=pd.merge(df_1,df_2,on='City',how='inner')
df_new
```

```python
df_new=pd.merge(df_1,df_2,on='City',how='outer')
df_new
```

```python
df_new=pd.merge(df_1,df_2,on='City',how='left')
df_new
```

```python
df_new=pd.merge(df_1,df_2,on='City',how='right')
df_new
```

```python
df_newdata=pd.DataFrame({"Temperature":[32,-99999,28,-99999,32,31,34],"WindSpeed":[6,7,-99999,7,-99999,2,5],"Event":['Rain','Sunny','Snow','0','Rain','Sunny','0']})
df_newdata
```

```python
df_newdata=df_newdata.replace([-99999,'0'],value=np.nan)
df_newdata
```

```python
df_newdata.index=[1,2,3,4,5,6,7]
df_newdata
```

Changing Index

```python
df_newdata.index=list(range(1,8))
df_newdata
```

```python
list(range(7))
```

```python
df_newdata=df_newdata.replace({'Temperature':np.nan,'WindSpeed':np.nan,'Event':np.nan},value='Null')
df_newdata
```

Replacing By Using Mapping

```python
df_newdata.replace({'Null':np.nan,'Rain':'Snow'},inplace=True)
df_newdata
```

Replacing using list

```python
df_newdata.replace(['Snow',np.nan],['Rain','Null'],inplace=True)
df_newdata
```

# Groupby

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.isna().sum() #To find the number of columns having missing values
```

```python
df.fillna({'Age':df['Age'].mean(),'Embarked':'S'},inplace=True)
df.isna().sum()
```

```python
df['Embarked'].value_counts()
```

```python
x=df.groupby('Embarked')
x.get_group('Q')[['Pclass','Fare','Ticket']]
```

```python
x.mean()
```

```python
x.get_group('Q')['Fare'].mean()
```

```python
x.mean()[['Age','Fare']]
```

```python
df['Pclass'].value_counts()
```

```python
y=df.groupby('Pclass')
c=y.get_group(3)
c
```

```python
b=y.get_group(2)
b
```
